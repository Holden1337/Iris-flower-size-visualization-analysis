# Using Pandas and Seaborn to visualize and analyze sample data about the size of iris flower leaves

```python
#import libraries
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#path to data
path_setosa = '/home/holden/programming/python/data_analysis/iris_setosa.csv'
path_versicolor = '/home/holden/programming/python/data_analysis/iris_versicolor.csv'
path_virginica = '/home/holden/programming/python/data_analysis/iris_virginica.csv'
path_iris = '/home/holden/programming/python/data_analysis/iris.csv'

#read data
set_data = pd.read_csv(path_setosa,index_col='Id')
ver_data = pd.read_csv(path_versicolor,index_col='Id')
vir_data = pd.read_csv(path_virginica,index_col='Id')
iris_data = pd.read_csv(path_iris,index_col='Id')
```

```python
#show first five rows of the data
set_data.head()
```

## Create a histogram to show the distribution of petal lengths for iris flowers

```python
#histogram
sns.distplot(a=iris_data['Petal Length (cm)'], kde=False)
```

## Creating a kernel density estimation plot to estimate the probability density function

```python
#kde plot
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

## Creating a 2D KDE plot 

```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde",shade = True)
```

## Plotting distributions of petal lenghts for different subspecies of the iris plant

```python
# Histograms for each species
sns.distplot(a=set_data['Petal Length (cm)'], label="Iris-setosa", kde=False)
sns.distplot(a=ver_data['Petal Length (cm)'], label="Iris-versicolor", kde=False)
sns.distplot(a=vir_data['Petal Length (cm)'], label="Iris-virginica", kde=False)

# Add title
plt.title("Histogram of Petal Lengths, by Species")

# Force legend to appear
plt.legend()
```

## creating KDE plots for different subspecies of the iris plants

```python
# KDE plots for each species
sns.kdeplot(data=set_data['Petal Length (cm)'], label="Iris-setosa", shade=True)
sns.kdeplot(data=ver_data['Petal Length (cm)'], label="Iris-versicolor", shade=True)
sns.kdeplot(data=vir_data['Petal Length (cm)'], label="Iris-virginica", shade=True)

# Add title
plt.title("Distribution of Petal Lengths, by Species")
```
